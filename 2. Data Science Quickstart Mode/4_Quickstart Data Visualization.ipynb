### 1. Some Basics

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
```

```python
y = x**2 + 3
```

```python
print(x,y)
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y)
plt.show()
```

### 2. Different Ploting Styles Available in Matplotlib

```python
plt.style.available
```

### 3. Using a Particular Style

```python
plt.style.use('seaborn')
plt.plot(x,y)
plt.show()
```

### 4. Giving a Title to the Plot

```python
plt.plot(x,y)
plt.title("My Line Plot")
plt.show()
```

### 5. Giving a Label - to tell what does the line represent
#### The label doesn't appear in the Plotted Graph unless legend() function is called.

```python
plt.plot(x,y,label = "Stock Price")
plt.title("My Line Plot")
plt.show()
```

```python
plt.plot(x,y,label = "Stock Price")
plt.title("My Line Plot")
plt.legend() # To make label vision in the Graph
plt.show()
```

### 6. Plotting more than 1 line in a single Graph.

```python
plt.plot(x, y, label = "Stock Price 1")
plt.plot(x, x**3, label = "Stock Price 2")
plt.legend()
plt.show()
```

### 7. Giving x-axis and y-axis label

```python
plt.plot(x,y, label = "Stock Price")
plt.title("My Line Plot")
plt.legend()
plt.xlabel("Time Stamp")
plt.ylabel("Price")
plt.show()
```

### 8. Using Scatter Plot

```python
plt.style.use('dark_background')
```

```python
plt.scatter(x,y, label="Stock Price 1")
plt.scatter(x, x**3 , label = " Stock Price 2")
plt.title("My Line Plot")
plt.show()
```
